```python
import nltk 
from nltk.translate import IBMModel1
from nltk.translate import AlignedSent 
from nltk.tokenize import word_tokenize
```

```python
bitext=[
    AlignedSent(["the ","house"],["la","mansion"]),
    AlignedSent(["the","book"],["la","livre"]),
    AlignedSent(["a","book"],["un","livre"]),
    AlignedSent(["my","book"],["mon","livre"]),
    AlignedSent(["my","house"],["mon","maison"])
    
]
```

```python
ibm1=IBMModel1(bitext,50)
sentence= input("enter the sectence:")
words=word_tokenize(sentence.lower())
translated=[]
for word in words:
    translations=ibm1.translation_table.get(word,{})
    best_word= max(translations,key=translations.get,default=word)
    translated.append(best_word)
print("\n Translated sentence:")
print(" ".join(translated))
```
