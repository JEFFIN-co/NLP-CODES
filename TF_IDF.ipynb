```python
import nltk 
from nltk import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
```

```python
stopword=set(stopwords.words('english'))
def preprocessing(text):
    words=word_tokenize(text.lower())
    check=[word for word in words if word not in stopword and word.isalpha()]
    return " ".join(check)
```

```python
n=int(input("enter the no of documents:"))
documents=[]
for i in range(n):
    doc=input(f"enter the documents{i+1}:")
    documents.append(preprocessing(doc))
```

```python
vectorizer=TfidfVectorizer()
tfidf_matrix=vectorizer.fit_transform(documents)
print('feature name(vocab)\n')
print(vectorizer.get_feature_names_out())
print('tf_idf matrix \n')
print(tfidf_matrix.toarray())
```
