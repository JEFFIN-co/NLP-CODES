```python
import nltk
from nltk import word_tokenize
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
filename='file.txt'
with open(filename,'r') as file:
    sentence=[line.strip() for line in file if line.strip()]
    if not sentence:
        print('error in reading the file ')
        exit(0)
    else:
        print("loaded",len(sentence),'sentence')
  
def preprocess(text):
    stopword=set(stopwords.words('english'))
    word=word_tokenize(text.lower())
    check=[w for w in word if w not in stopword and  w.isalpha()]
    return ' '.join(check)
def similarity(sentance,input_text):
    corpus=sentence+[input_text]
    vectorizer=TfidfVectorizer()
    tfidf_matrix=vectorizer.fit_transform(corpus)
    similar=cosine_similarity(tfidf_matrix[-1],tfidf_matrix[:-1])
    index=similar.argmax()
    score=similar[0][index]
    if score==0:
        print("i don't understand")
    return sentence[index]
print("the bot is ready to answer")
while True:
    input_text=input('Enter the question:')
    if (input_text.lower() == 'exit'):
          print('bot:thankyou')
          break
   
    prepocessed_input_text=(preprocess(input_text))
    result=similarity(sentence,prepocessed_input_text)
    print('bot:',result)
```
