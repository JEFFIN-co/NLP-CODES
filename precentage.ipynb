```python
from nltk import word_tokenize, FreqDist
```

```python
text=input('enter the text')
word=word_tokenize(text)
print(word)
```

```python
freq=FreqDist(word)
print(freq.most_common())
```

```python
count=len(word)
for i in word :
    print('the precentage of ', i , 'is :',str(round(freq[i]/count*100,2)))
```
