```python
import nltk
from nltk import pos_tag,ne_chunk, word_tokenize as wd
```

```python
sentence=input('enter the text:')
words=wd(sentence)
postag=pos_tag(words)
named_entities=ne_chunk(postag)
```

```python
for chunk in named_entities:
    if hasattr(chunk, 'label'):
        entity_name= ' '.join(c[0] for c in chunk.leaves())
        entity_type=chunk.label()
        print(f"{entity_name} -> {entity_type}")
```
