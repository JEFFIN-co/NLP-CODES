```python
import nltk 
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity 
from nltk.corpus import stopwords
from nltk import word_tokenize , sent_tokenize
```

```python
stopword=set(stopwords.words('english'))
def preprocessed(text):
    sent=sent_tokenize(text.lower())
    for s in sent :
        words=word_tokenize(s)
        check=[word for word in words if word not in stopword and word.isalpha()]
        return" ".join(check)
    
```

```python
def most_similar_sentence(filename,input_sentence):
    with open(filename, 'r') as file :
        sentence=[line.strip() for line in file if line.strip()]

    corpus=sentence+[input_sentence]
    vectorizer=TfidfVectorizer()
    tf_idf_matrix=vectorizer.fit_transform(corpus)
    similarity_score=cosine_similarity(tf_idf_matrix[-1],tf_idf_matrix[:-1])
    similarity_index=similarity_score.argmax()

    return sentence[similarity_index],similarity_score
    
```

```python
text=input('enter the text :')
preprocess_text=preprocessed(text)
filename='text.txt'
result, score=most_similar_sentence(filename,preprocess_text)
print('the similar sentence',result)
print('score:',score.max())
print('the context:')
with open (filename,'r') as file :
    print(file.read())
```
