```python
import nltk 
from nltk import word_tokenize 
from nltk.corpus import stopwords
```

```python
text= input('Enter the sentence:')
words=word_tokenize(text)
stopword=set(stopwords.words('english'))
check=[word for word in words if word not in stopword]
```

```python
vocab=sorted(set(list(check)))
print('vocabulary',vocab)
print('OneHotEncoding\n')
print('word \t\t vector')
for v in vocab :
    vector=[1 if v==word else 0 for word in words]
    print( v ,':\t\t',vector)
```
